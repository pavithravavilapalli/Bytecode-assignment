5. Customer Analysis Using Sets
- january_customers = {101,102,103,104,105}
- february_customers = {103,104,105,106,107}
- Find:
- Repeat customers
- New customers in February
- Customers who purchased only in January



```python
january_customers = {101,102,103,104,105}
february_customers = {103,104,105,106,107}
repeat_customers = january_customers.intersection(february_customers)
print(repeat_customers)
new_customers = january_customers.difference(february_customers)
print(new_customers)
customers_only_january=january_customers.difference(february_customers)
print(customers_only_january)
```

6. Marketing Campaign Analysis
- Campaign A: A = {1001,1002,1003,1004,1005}
- Campaign B: B = {1004,1005,1006,1007}
- Find:
- Customers reached by both campaigns
-	Customers reached by either campaign
-	Customers reached by exactly one campaign


```python
A = {1001,1002,1003,1004,1004,1005}
B = {1004,1005,1006,1007}
reached_by_both = A.intersection(B)
print(reached_by_both)
reached_by_either = A.union(B)
print(reached_by_either)
reached_by_exactly_one = A.symmetric_difference(B)
print(reached_by_exactly_one)
```

12. Common Products Sold
- store1 = {"Laptop","Mouse","Keyboard","Monitor"}
- store2 = {"Mouse","Monitor","Printer","Scanner"}
- Find:
-	Common products
-	Exclusive products for each store
-	Total unique products


```python
store1 = {"Laptop","Mouse","Keyboard","Monitor"}
store2 = {"Mouse","Monitor","Printer","Scanner"}
print("common products:",store1.intersection(store2))
print("exclusive products for store1:",store1.difference(store2))
print("total unique products:",store1.union(store2))
```

13. Mini Case Study
- An e-commerce company ran campaigns in 3 months:

- jan = {101,102,103,104,105}
- feb = {103,104,105,106,107}
- mar = {105,106,107,108,109}

- Answer:

-	Total unique customers targeted.
-	Customers targeted in all months.
-	Customers targeted only once.
-	Customers targeted in January and February but not March.
-	Customers targeted in February and March but not January.



```python
jan = {101,102,103,104,105}

feb = {103,104,105,106,107}

mar = {105,106,107,108,109}
print("total unique customers targeted:",jan.union(feb,mar))
print("customers targeted in all months:",jan.intersection(feb,mar))
print("customers targeted only once:",jan.intersection(feb,mar))

#print("customers targted only once:",jan.symmetric_difference(feb))
a=jan.symmetric_difference(feb)
b=a.symmetric_difference(mar)
print("Customers targeted only once: ",b)
c=jan.intersection(feb)
d=feb.intersection(mar)
e=c.difference(d)
print("Customers targeted in January and February but not March: ",e)
f=feb.intersection(mar)
g=mar.intersection(jan)
h=f.difference(g)
print("Customers targeted in February and March but not January: ",h)
```
