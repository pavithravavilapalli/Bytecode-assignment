# Question 1
- data = ["100", "200", "300", "400"]
- Convert all values to integers and find:
- Total sum
- Average
- Maximum value
- Minimum value

```python
data = ["100", "200", "300", "400"]
sum=0
for i in range(len(data)):
  a=int(data[i])
  print("elements:",a)
  sum=sum+a
  avg=sum/len(data)
  maxx=max(data)
  minn=min(data)
print("Sum:",sum)
print("Avg:",avg)
print("Max:",maxx)
print("Min:",minn)


```

#Challange 2
- students = {
"Ram":[80,90,85],
"Priya":[95,88,92],
"Hari":[70,75,80]
}
- Find:
- Total marks of each student
- Average marks of each student
- Student with highest average


```python
students = {
    "Ram": [80, 90, 85],
    "Priya": [95, 88, 92],
    "Hari": [70, 75, 80]
}

student_averages = {}

for student_name, marks_list in students.items():
    total_marks = sum(marks_list)
    average_marks = total_marks / len(marks_list)
    student_averages[student_name] = average_marks

    print(f"\nStudent: {student_name}")
    print(f"  Total Marks: {total_marks}")
    print(f"  Average Marks: {average_marks:.2f}")


if student_averages:
    highest_avg_student = max(student_averages, key=student_averages.get)
    highest_average = student_averages[highest_avg_student]

    print(f"Student with the highest average: {highest_avg_student} (Average: {highest_average:.2f})")
else:
    print("\nNo student data to calculate averages.")
```
