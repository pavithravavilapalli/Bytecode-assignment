21.	Build Frequency Dictionary
- products = [“Laptop”, “Mobile”, “Laptop”, “Tablet”, “Mobile”, “Laptop”]
- Create a frequency dictionary using a while loop.


```python
products=["Laptop","Mobile","Laptop","Tablet","Mobile","Laptop"]
freq={}
i=0
while i<len(products):
  product=products[i]
  if product in freq:
    freq[product]=freq[product]+1
  else:
    freq[product]=1
  i=i+1
print(freq)

```

30.	Sales Trend Analysis
- sales = [1000, 1200, 1100, 1500, 1400]
- Count:
-	Number of increasing months
-	Number of decreasing months.


```python
sales = [1000, 1200, 1100, 1500, 1400]
i=0
increasing_months=0
decreasing_months=0
while i < len(sales):
  if sales[i] > sales[i-1]:
    increasing_months=increasing_months+1
  elif sales[i] < sales[i-1]:
    decreasing_months=decreasing_months+1
  i=i+1
print("Number of increasing months :",increasing_months)
print("Number of decreasing months :",decreasing_months)
```
