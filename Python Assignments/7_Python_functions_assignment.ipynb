#1. Write a function that returns the largest number from a list without using max().

```python
def maxx(lst):
  max_num = lst[0]
  for num in lst:
    if num > max_num:
      max_num = num
  return max_num
maxx([10,20,30,40])
```

#4. Find Second Highest Number

```python
def second_highest(lst):
    unique = []
    for i in lst:
        if i not in unique:
            unique.append(i)
    unique.sort()
    return unique[-2]
second_highest([10,20,30,40])
```

#5. Find the Separate Even and Odd Numbers : Return two lists.

```python
def even_odd(lst):
    even = []
    odd = []
    for i in lst:
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)
    return even, odd
even_odd([1,2,3,4,5,6,7,8,9])
```

#6. Find Duplicate Elements : ls = [1,2,3,2,4,5,1]

```python
def duplicates(ls):
    dup = []
    for i in ls:
        if ls.count(i) > 1 and i not in dup:
            dup.append(i)
    return dup
duplicates([1,2,3,2,4,5,1])
```

#10. Reverse Each Word (not string): input: I love python

```python
def reverse(sentence):
    words = sentence.split()
    result = ""
    for i in words:
        result = result + i[::-1] + " "
    return result.strip()
reverse("I love python")
```

#17. Most Frequent Word

```python
def most_frequent_word(text):
    words = text.split()
    frequency = {}
    for word in words:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    highest = 0
    result = ""
    for word in frequency:
        if frequency[word] > highest:
            highest = frequency[word]
            result = word
    return result
text = "python sql python excel sql python"
most_frequent_word(text)
```

#27. Find Pairs Whose Sum = Target
- Input:
- nums = [2,7,11,15,3,6]
- target = 9


```python
def pair_sum(numbers, target):
    pairs = []
    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] + numbers[j] == target:
                pairs.append((numbers[i], numbers[j]))
    return pairs
pair_sum([2,7,11,15,3,6], 9)
```

#28. Compress String
- Input: aaabbccdddd


```python
def compress(text):
    result = ""
    count = 1
    for i in range(1, len(text)):
        if text[i] == text[i-1]:
            count = count + 1
        else:
            result = result + text[i-1] + str(count)
            count = 1
    result = result + text[-1] + str(count)
    return result
compress("aaabbccdddd")
```

#29. Expand String
- Input: a3b2c2


```python
def expand(text):
    result = ""
    for i in range(0, len(text), 2):
        character = text[i]
        count = int(text[i+1])
        result = result + character * count
    return result
expand("a3b2c2")
```

#30. Find Consecutive Duplicate Count
- Input: [1,1,1,2,2,3,3,3,3]


```python
def consecutive_count(numbers):
    result = {}
    current = numbers[0]
    count = 1
    for i in range(1, len(numbers)):
        if numbers[i] == current:
            count = count + 1
        else:
            result[current] = count
            current = numbers[i]
            count = 1
    result[current] = count
    return result
numbers = [1,1,1,2,2,3,3,3,3]
consecutive_count(numbers)
```

#31. Nested Dictionary Analysis
- sales = {
    "North":{
        "Jan":1000,
        "Feb":2000
    },
    "South":{
        "Jan":3000,
        "Feb":1500
    }
}


```python
def nested_sales_analysis(sales):
    grand_total = 0
    for region in sales:
        region_total = 0
        for month in sales[region]:
            region_total = region_total + sales[region][month]
        print(region, "Total =", region_total)
        grand_total = grand_total + region_total
    print("Grand Total =", grand_total)
sales = {
    "North":{
        "Jan":1000,
        "Feb":2000
    },
    "South":{
        "Jan":3000,
        "Feb":1500
    }
}
nested_sales_analysis(sales)
```

#32. Customer Lifetime Value
- customers = {
"John":[1000,2000,1500],
"Priya":[5000,3000],
"David":[1000]
}


```python
def customer_lifetime_value(customers):
    totals = {}
    highest_customer = ""
    highest_spend = 0
    total_spend = 0
    for customer in customers:
        customer_total = 0
        for amount in customers[customer]:
            customer_total = customer_total + amount
        totals[customer] = customer_total
        total_spend = total_spend + customer_total
        if customer_total > highest_spend:
            highest_spend = customer_total
            highest_customer = customer
    average_spend = total_spend / len(customers)
    print("Customer Totals =", totals)
    print("Highest Customer =", highest_customer)
    print("Average Spend =", average_spend)
customers = {
    "John":[1000,2000,1500],
    "Priya":[5000,3000],
    "David":[1000]
}
customer_lifetime_value(customers)
```

#33. Build a Mini Dashboard Function
- Input:
- sales = [1000,2000,5000,3000,7000]
- Return:
- Total Sales
- Average Sales
- Highest Sales
- Lowest Sales
- Sales Count


```python
def dashboard(sales):
    total = 0
    highest = sales[0]
    lowest = sales[0]
    count = len(sales)
    for sale in sales:
        total = total + sale
        if sale > highest:
            highest = sale
        if sale < lowest:
            lowest = sale
    average = total / count
    result = {
        "Total": total,
        "Average": average,
        "Highest": highest,
        "Lowest": lowest,
        "Count": count
    }
    return result
sales = [1000,2000,5000,3000,7000]
dashboard(sales)
```

#34. Validate Password
- Rules:
- Minimum 8 chars
- At least 1 uppercase
- At least 1 lowercase
- At least 1 digit
- Return: Valid Or Invalid


```python
def validate_password(password):
    if len(password) < 8:
        return "Invalid"
    has_upper = False
    has_lower = False
    has_digit = False
    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True
    if has_upper and has_lower and has_digit:
        return "Valid"
    return "Invalid"
validate_password("Python123")
```
